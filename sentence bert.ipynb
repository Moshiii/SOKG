```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Biker_1k_rdm.csv')
df=df[["title","answer"]]
```

```python
train_title_list = df["title"].to_list()
```

```python
train_embedding = []
for i in range(len(train_title_list)):
    sentence = train_title_list[i]
    embedding = model.encode(sentence, convert_to_tensor=True)
    if i%1000==0:
        print(i) 

    train_embedding.append(embedding.tolist())
```

```python
df["embedding"]=train_embedding
```

```python
df.to_csv('Biker_1k_rdm.embed.csv')
```
