```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def remove_prefix(text, prefix):
    if text.startswith(prefix):
        return text[len(prefix):]
    else:
        return text
    
def remove_suffix(text, prefix):
    if text.endswith(prefix):
        return text[:-len(prefix)]
    else:
        return text
```
